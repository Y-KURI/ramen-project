```python
import numpy as np
import pandas as pd
import os
from PIL import Image

os.chdir("F:")
csv = pd.read_csv("ramen.csv", encoding = "utf=8")
df = pd.DataFrame(csv)
#シャッフルする
df = df.reindex(np.random.permutation(df.index))
img_num = 500

#らーめん4種のデータに分ける
miso = df[df["style"]=="[ラーメン/味噌]"][:img_num]
sio = df[df["style"]=="[ラーメン/塩]"][:img_num]
tonkotsu = df[df["style"]=="[ラーメン/豚骨]"][:img_num]
shouyu = df[df["style"]=="[ラーメン/醤油]"][:img_num]

#4種のＤＦと対応する画像保存先フォルダを作成
df4 = [miso, sio, tonkotsu, shouyu]
style4 = ["miso", "sio", "tonkotsu", "shouyu"]


dir_from =  "F:/ramen_image/"
dir_into = "F:/ramen_img100/"


def save_img(df4, style4):
    #画像の保存先を4つ作成、４つのディレクトリを切り替え
    os.chdir(dir_into)
    for i in range(len(style4)):
        try:
            os.mkdir("./" + style4[i])
            
        except FileExistsError:
            pass
        dir = dir_into + style4[i]
        os.chdir(dir)
        print(dir)
        
        #４つのＤＦ内のIDカラムより、らーめん４種の画像ＩＤを取得、画像をpngでコピーする
        ID = df4[i].ID
        for id in ID:
            id = str(id)
            img = Image.open(dir_from + id + ".jpg")
            copy_img = img.copy()
            copy_img.save(id + ".png")
            
        os.chdir(dir_into)
        


save_img(df4, style4)


```
