```python
#スコアの度数、グラフ

import pandas as pd
csv = pd.read_csv("ramen.csv", encoding = "utf=8")
df = pd.DataFrame(csv)

#得点の範囲を調べて、60～69、70～79、80～89、90～100のように階級分けしたい

#df.sort(columns="score")# ソートして最大最小を調べる、Nanの次は100、終わりが99点に？？得点を上手くソート出来ない、、

#objectからintに変換を試みる
#①df = df["score"].str.extract('([0-9])').fillna(0).astype(int)   #数字だけにしてタイプ変換 dtype: int32に変換できるが一桁数字に??
#②df = df["score"].convert_objects(convert_numeric= True)  #Name: score, dtype: object 変化無し
#③df = pd.to_numeric(df["score"])#Unable to parse string "82点" at position 0


score = df['score'].dropna()
score = pd.Series(score)
score_freq = score.value_counts().sort_index()

import matplotlib.pyplot as plt

score_freq.plot(kind='bar')# 度数分布でplotしてみる
plt.show()
```

```python
#スコアの度数分布、多い順

import pandas as pd
csv = pd.read_csv("ramen.csv", encoding = "utf=8")
df = pd.DataFrame(csv)


score = df['score'].dropna()
score = pd.Series(score)
score_freq = score.value_counts()#.sort_index

import matplotlib.pyplot as plt

score_freq.plot(kind='bar')# 度数分布でplotしてみる
plt.show()
```
