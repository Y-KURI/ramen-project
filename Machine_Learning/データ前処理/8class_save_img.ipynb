```python
import numpy as np
import pandas as pd
import os
from PIL import Image

#os.chdir("F:")
csv = pd.read_csv("ramen.csv", encoding = "utf=8")
df = pd.DataFrame(csv)
#シャッフルする
df = df.reindex(np.random.permutation(df.index))

#らーめん階級分けする
#scoreの値をintに変換
df = df[df.score.apply(lambda x: isinstance(x, int))]
nb_data = 100

df1 = df.query(' score < 65')[:nb_data]
df2 = df.query('65 <= score < 70')[:nb_data]
df3 = df.query('70 <= score < 75')[:nb_data]
df4 = df.query('75 <= score < 80')[:nb_data]
df5 = df.query('80 <= score < 85')[:nb_data]
df6 = df.query('85 <= score < 90')[:nb_data]
df7 = df.query('90 <= score < 95')[:nb_data]
df8 = df.query('95 <= score <= 100')[:nb_data]



#4種のＤＦと対応する画像保存先フォルダを作成
df_list = [df1, df2, df3, df4, df5, df6, df7, df8]
class8 = ["class1", "class2", "class3", "class4", "class5", "class6", "class7", "class8"]


dir_from =  "F:/ramen_image/"
dir_into = "F:/ramen_img100/"


def save_img(df_list, classes):
    #画像の保存先を作成、ディレクトリを切り替え
    os.chdir(dir_into)
    for i in range(len(classes)):
        try:
            os.mkdir("./" + classes[i])
            
        except FileExistsError:
            pass
        dir = dir_into + classes[i]
        os.chdir(dir)
        print(dir)
        
        #ＤＦ内のIDカラムより、画像ＩＤを取得、画像をpngでコピーする
        ID = df_list[i].ID
        for id in ID:
            id = str(id)
            img = Image.open(dir_from + id + ".jpg")
            copy_img = img.copy()
            copy_img.save(id + ".png")
        os.chdir(dir_into)#ディレクトリ戻る
```

```python
save_img(df_list, class8)
```
